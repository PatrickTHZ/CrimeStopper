```python
import pandas as pd
```

```python
df = pd.read_csv('SuburbData25Q4.csv')
```

```python
display(df)
```

```python
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

# Ensure plots show inside notebook
%matplotlib inline

# Extract month columns
month_cols = [col for col in df.columns if "19" in col or "20" in col]

# Dropdown widgets
suburb_dropdown = widgets.Dropdown(
    options=sorted(df["Suburb"].unique()),
    description="Suburb:",
    layout=widgets.Layout(width="300px")
)

offence_dropdown = widgets.Dropdown(
    options=["All offences"] + sorted(df["Offence category"].unique()),
    description="Offence:",
    layout=widgets.Layout(width="300px")
)

# Plotting function
def plot_crime(suburb, offence):
    plt.figure(figsize=(14,6))

    # Filter by suburb
    df_sub = df[df["Suburb"] == suburb]

    # If user selects ALL offences
    if offence == "All offences":
        ts = df_sub[month_cols].sum(axis=0)
    else:
        df_sub = df_sub[df_sub["Offence category"] == offence]
        ts = df_sub[month_cols].sum(axis=0)

    # Convert index to datetime
    ts.index = pd.to_datetime(ts.index, format="%b %Y")

    # Plot
    plt.plot(ts.index, ts.values, linewidth=2, marker="o")
    plt.title(f"{offence} in {suburb} (1995–2002)")
    plt.xlabel("Year")
    plt.ylabel("Offences")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Interactive UI
widgets.interact(
    plot_crime,
    suburb=suburb_dropdown,
    offence=offence_dropdown
)
```
